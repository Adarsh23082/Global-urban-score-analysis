## Top 10 urban areas by overall score

```python
data_copy = data.copy()
# Select columns from 'Housing' to 'Outdoors'
score_columns = data_copy.columns[3:]
data_copy['Overall_Score'] = data_copy[score_columns].sum(axis=1)
top_10_urban_areas=data_copy.sort_values(by='Overall_Score', ascending=False)[['UA_Name', 'Overall_Score']].head(10)
top_10_urban_areas
```

```python
plt.barh(top_10_urban_areas['UA_Name'], top_10_urban_areas['Overall_Score'], color='skyblue')
plt.xlabel('Overall Score')
plt.ylabel('Urban Area')
plt.title('Top 10 Urban Areas by Overall Score')
plt.show()
```

## Scores of urban areas vary by continent

```python
mean_scores_by_continent = data.groupby('UA_Continent')[score_columns].mean().reset_index()
mean_scores_by_continent 
```

```python
mean_scores_by_continent.set_index('UA_Continent').T.plot(kind='bar', figsize=(15, 10))
plt.title('Mean Scores of Urban Areas by Continent')
plt.ylabel('Mean Score')
plt.xlabel('Category')
plt.legend(title='Continent', loc='upper right')
plt.show()
```

## Top 10 Urban Areas in Asia

```python
asia_data = data_copy[data_copy['UA_Continent'] == 'Asia'].copy()
asia_data['Overall_Score2'] = asia_data.iloc[:, 3:].sum(axis=1)
top_10_urban_areas_inAsia = asia_data.sort_values(by='Overall_Score2', ascending=False)[['UA_Name', 'Overall_Score2']].head(10)
top_10_urban_areas_inAsia
```

## Scores of urban areas in Europe compare to those in North America across various categories

```python
europe_vs_north_america = data[data['UA_Continent'].isin(['Europe', 'North America'])]
mean_scores_comparison = europe_vs_north_america.groupby('UA_Continent')[score_columns].mean().reset_index()
mean_scores_comparison
```

```python
mean_scores_comparison.set_index('UA_Continent').T.plot(kind='bar', figsize=(15, 10))
plt.title('Mean Scores: Europe vs North America')
plt.ylabel('Mean Score')
plt.xlabel('Category')
plt.legend(title='Continent', loc='upper left')
plt.show()
```

## Urban Areas with the Highest Growth Potential (Business Freedom, Venture Capital, and Startup Scores)

```python
data['Growth_Potential'] = data[['Business Freedom', 'Venture Capital', 'Startups']].mean(axis=1)
top_10_growth_potential = data[['UA_Name', 'Growth_Potential']].sort_values(by='Growth_Potential', ascending=False).head(10)
top_10_growth_potential
```

```python
sns.barplot(x='Growth_Potential', y='UA_Name', data=top_10_growth_potential)
plt.title('Top 10 Urban Areas with the Highest Growth Potential(Business Freedom, Venture Capital, and Startup Scores)')
plt.xlabel('Growth Potential Score')
plt.ylabel('Urban Area')
plt.show()
```

## Top Urban Areas for each attributes 

```python
highest_urban_areas = {col: data.loc[data[col].idxmax()]['UA_Name'] for col in data.columns[3:]}
highest_urban_areas
```

```python
highest_values = data[data.columns[3:]].max()
highest_values
```

```python

# Find the urban areas corresponding to the highest values
highest_urban_areas = {}
for category in highest_values.index:
    urban_area = data.loc[data[category].idxmax(), 'UA_Name']
    highest_urban_areas[category] = urban_area

# Create a DataFrame for plotting
highest_values_df = highest_values.reset_index()
highest_values_df.columns = ['Category', 'Value']
highest_values_df['Urban Area'] = highest_values_df['Category'].map(highest_urban_areas)

# Plot the bar graph
plt.figure(figsize=(12, 8))
sns.barplot(data=highest_values_df, x='Category', y='Value', hue='Urban Area')
plt.title('Highest Value of Each Category by Urban Area')
plt.ylabel('Highest Value')
plt.xlabel('Category')
plt.xticks(rotation=90)
plt.legend(title='Urban Area', loc='upper left')
# plt.tight_layout()
plt.show()
```
